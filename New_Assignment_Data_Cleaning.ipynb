```python
# Define a function that will categorize the smokinghabit into disctinct categories
def categorize_smoking(habit):
    if 'Active Smoker' in habit:
        return 'Active Smoker'
    elif 'Former Smoker' in habit:
        return 'Former Smoker'
    elif 'No Smoker' in habit:
        return 'Non-Smoker'
    else:
        return 'Former Smoker' # 'Unknown' or any other unexpected values are ascribed as Former smokers using the modal score
```

```python
# Defining a function that will categorize the numfamlung values into disctinct categories

def upd_numfamlung(numfamlung):
    if pd.isna(numfamlung):
        return '0'

    num = int(numfamlung)

    if num in [1, 2, 3, 4, 5]:
        return str(num)
    else:
        return '0' # 'Unknown' or any other unexpected values are defined as 'No' as the other factors adds up to a No
```

```python
##>>>>>>>>> Initial Stage of Detection (inistage) column
```

```python
## Inspecting unique values in inistage column

selected_data['inistage'].value_counts()
```

```python
### Categorize the initial stage values into four elements: (I, II, III, IV)

def upd_inistage(inistage):
    if pd.isna(inistage):
        return 'Unknown'

    inistage = str(inistage).upper() # to convert all values to capital letters

    # Stage IV
    if 'IVB' in inistage or 'IVA' in inistage or 'IV' in inistage:
        return 'IV'

    # Stage III
    elif 'IIIC' in inistage or 'IIIB' in inistage or 'IIIA' in inistage:
        return 'III'

    # Stage II
    elif 'IIB' in inistage or 'IIA' in inistage:
        return 'II'

    # Stage I
    elif (
        'IA1' in inistage or
        'IA2' in inistage or
        'IA' in inistage or
        'IB' in inistage or
        'LIMITED' in inistage
    ):
        return 'I'

    # Other mappings
    elif 'EXTENDED' in inistage or 'OTHER' in inistage:
        return 'IV'

    else:
        return 'Unknown'
```

```python
# Applying the defined categorization function to create a new column
selected_data['upd_inistage'] = selected_data['inistage'].apply(upd_inistage)

print("Unique values and their counts in the new 'upd_inistage' column:")
print(selected_data['upd_inistage'].value_counts())


# Display the first few rows with the new column

display(selected_data[['upd_inistage']].head())
```

```python
## Defining the responses in the Treatment Response (response_pharmtreat1_CUI) column

# Distinct representation of each value
print(selected_data['response_pharmtreat1_CUI'].value_counts())
```

```python
## To categorize the column with defined value using the function below

def upd_TreatResponse(response_cui_value):
    if pd.isna(response_cui_value):
        return 'Unknown'
    elif response_cui_value == 'EE - C0677946':
        return 'Stable Disease'
    elif response_cui_value == 'NE - C1883425':
        return 'Not Evaluable'
    elif response_cui_value == 'NR - C0445106':
        return 'No Referral'
    elif response_cui_value == 'PR - C0242656':
        return 'Disease Progression'
    elif response_cui_value == 'RC - C1275810':
        return 'Complete Response'
    elif response_cui_value == 'RP - C1275833':
        return 'Partial Response'
    else:
        return 'Unknown'
```

```python
# Applying this function to create a new column (updated_response_pharmtreat1_CUI)

selected_data['updated_response_pharmtreat1_CUI'] = selected_data['response_pharmtreat1_CUI'].apply(upd_TreatResponse)

print("Unique values and their counts in the new 'updated_response_pharmtreat1_CUI' column:")
print(selected_data['updated_response_pharmtreat1_CUI'].value_counts())

display(selected_data[['response_pharmtreat1_CUI', 'updated_response_pharmtreat1_CUI']].head())
```

```python

#>>>>>>>>>> To Determine Valid Rows for a Clean Table
```

```python
''' Since treatment response is the target variable, records lacking both
treatment initiation evidence and response should be excluded, while
records with confirmed treatment initiation and observed response form
the valid analytical cohort.'''
```

```python
''' Logic to determine valid rows
TR IS NOT NULL
AND
TD IS NULL or  TT IS NOT NULL'''
```

```python

## To achieve the above, set the following column as binary (phartreat1, phartreatstartdate1, and response_pharmtreat1_CUI) ...
```

```python
# Create a new column 'phartreat1_binary' where 1 indicates data and 0 indicates empty (NaN)

selected_data['phartreat1_binary'] = selected_data['phartreat1'].notnull().astype(int)

print("Value counts for the new 'phartreat1_binary' column:")
print(selected_data['phartreat1_binary'].value_counts())

display(selected_data[['phartreat1', 'phartreat1_binary']].head())
```

```python
# Create a new column for Treatment date 'phartreatstartdate1_binary' where 1 indicates data and 0 indicates empty (NaN)

selected_data['phartreatstartdate1_binary'] = selected_data['phartreatstartdate1'].notnull().astype(int)

print("Value counts for the new 'phartreatstartdate1_binary' column:")
print(selected_data['phartreatstartdate1_binary'].value_counts())

display(selected_data[['phartreatstartdate1', 'phartreatstartdate1_binary']].head())
```

```python
# Create a new column 'response_pharmtreat1_CUI_binary' where 1 indicates data and 0 indicates empty (NaN)
selected_data['response_pharmtreat1_CUI_binary'] = selected_data['response_pharmtreat1_CUI'].notnull().astype(int)

print("Value counts for the new 'response_pharmtreat1_CUI_binary' column:")
print(selected_data['response_pharmtreat1_CUI_binary'].value_counts())

display(selected_data[['response_pharmtreat1_CUI', 'response_pharmtreat1_CUI_binary']].head())
```

```python
''' Based on this logic, we will define 'valid rows' where:

Treatment Response (TR) is NOT NULL (i.e., response_pharmtreat1_CUI_binary is 1)
AND
(Treatment Date (TD) IS NULL (i.e., phartreatstartdate1_binary is 0)
OR
Treatment Type (TT) IS NOT NULL (i.e., phartreat1_binary is 1))
'''
```

```python
# Applying the logic to determine valid_rows of data
valid_data = selected_data[
    (selected_data['response_pharmtreat1_CUI_binary'] == 1) &
    (
        (selected_data['phartreatstartdate1_binary'] == 0) |
        (selected_data['phartreat1_binary'] == 1)
    )
]

print(f"Number of valid rows of data: {len(valid_data)}")
print("First 5 valid rows of data:")
display(valid_data.head())
```

```python
# Observations were excluded where treatmentresponse and treatment dat is missing. This surges that tno treatment was possibly done and the target variable was undefined. On the other hand, the logic preserves the final curated dataset and the criteria here is that the treatment response was present and there is the evidence of a treatment inititiation which is implied by either the date or treatment type. As such it infers that the treatment response is not null and one of either the treatment date or treatment type is not null.
```
